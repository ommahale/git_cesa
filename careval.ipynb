```python
import pandas as pd 
import sklearn
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np 
from sklearn import linear_model,preprocessing
```

```python
df=pd.read_csv("car_evaluation.csv")
le=preprocessing.LabelEncoder()
buying=le.fit_transform(list(df['buying'])) 
maint=le.fit_transform(list(df['maint'])) 
lug_boot=le.fit_transform(list(df['lug_boot'])) 
safety=le.fit_transform(list(df['safety'])) 
cls=le.fit_transform(list(df['class'])) 
door=le.fit_transform(list(df['door'])) 
persons=le.fit_transform(list(df['persons']))
 
```

```python
predict="class"
X=list(zip(buying,maint,lug_boot,safety,door,persons))
Y=list(cls)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1)
```

```python
model= KNeighborsClassifier(n_neighbors=9)
model.fit(X_train,Y_train)
acc=model.score(X_test,Y_test)
print(acc)
predicted=model.predict(X_test)
names=["unacc","acc","good","vgood"]
for x in range(len(X_test)):
    if predicted[x]==0:
        print(names[0])
    elif predicted[x]==1:
        print(names[1])
    elif predicted[x]==2:
        print(names[2])
    elif predicted[x]==3:
        print(names[3])
```
